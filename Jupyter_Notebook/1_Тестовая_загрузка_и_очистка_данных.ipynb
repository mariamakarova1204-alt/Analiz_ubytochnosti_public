```python
#присваиваем названия нашим колонкам
df.columns = [
    'Дата',
    'Документ',
    'Операции',
    'Счет_дебет',
    'Сумма_дебет',
    'Счет_кредит',
    'Сумма_кредит',
    'Текущее_сальдо',
    'Сальдо'
]
```

```python
#для анализа нужна колонка Сумма_кредит. Смотрим тип колонки
df['Сумма_кредит'].head()
```

```python
#переводим str в float64
cols = ['Сумма_дебет', 'Сумма_кредит', 'Текущее_сальдо', 'Сальдо']
for col in cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(' ', '', regex=False)
        .str.replace(',', '.', regex=False)
    )

df[cols] = df[cols].apply(
    pd.to_numeric,
    errors='coerce'
)
```
